# üîÑ Codificaci√≥n de Variables Categ√≥ricas  

Este c√≥digo utiliza `LabelEncoder` para convertir variables categ√≥ricas en valores num√©ricos, facilitando su uso en modelos de Machine Learning.  

## üìù C√≥digo:

````python
from sklearn.preprocessing import LabelEncoder

# Crear una instancia de LabelEncoder
le = LabelEncoder()

# Codificar las variables categ√≥ricas de tipo 'object' (como 'Gender', 'family_history_with_overweight', etc.)
categorical_columns = ['Gender', 'family_history_with_overweight', 'FAVC', 'CAEC', 'SMOKE', 'SCC', 'CALC', 'MTRANS']

for col in categorical_columns:
    datos[col] = le.fit_transform(datos[col])

# Codificar la columna 'NObeyesdad' (variable objetivo) a n√∫meros
datos['NObeyesdad'] = le.fit_transform(datos['NObeyesdad'])
````

# ‚úÇÔ∏è Separaci√≥n de Variables: Caracter√≠sticas (X) y Objetivo (Y)

En este paso, dividimos el conjunto de datos en **variables predictoras (X)** y **variable objetivo (Y)**.

## üìù C√≥digo:

````python
X = datos.drop('NObeyesdad', axis=1)
Y = datos['NObeyesdad']
````

# üîÄ Divisi√≥n del Conjunto de Datos: Entrenamiento y Prueba

En este paso, se divide el dataset en dos partes:  
- **80%** para entrenamiento (X_entrena, Y_entrena) üìö  
- **20%** para prueba (X_prueba, Y_prueba) üß™  

## üìù C√≥digo:

````python
from sklearn.model_selection import train_test_split
X_entrena, X_prueba, Y_entrena, Y_prueba = train_test_split(X, Y, test_size=0.2, random_state=42)
````

# üå≤ Entrenamiento del Modelo con Random Forest

Se utiliza **Random Forest**, un modelo de clasificaci√≥n basado en m√∫ltiples √°rboles de decisi√≥n, para predecir el nivel de obesidad.

üìå Analog√≠a:
Puedes imaginar Random Forest como una votaci√≥n en la que 100 personas (√°rboles de decisi√≥n) dan su opini√≥n sobre el resultado.

üîπ Ejemplo: Si est√°s decidiendo qu√© pel√≠cula ver y le preguntas a 100 amigos, algunos dir√°n "acci√≥n", otros "comedia", pero la decisi√≥n final ser√° la que tenga m√°s votos.

‚úÖ En el modelo, cada √°rbol hace una predicci√≥n y la decisi√≥n final se toma por mayor√≠a de votos (clasificaci√≥n) o por promedio (regresi√≥n).

## üìù C√≥digo:

````python
from sklearn.ensemble import RandomForestClassifier

modelo_rf = RandomForestClassifier(n_estimators=100, random_state=42)

modelo_rf.fit(X_entrena, Y_entrena)
````

## üìä Validaci√≥n Cruzada con `cross_val_score`

Para evaluar la precisi√≥n del modelo, utilizamos **validaci√≥n cruzada** con `cross_val_score`.

ü§î ¬øQu√© significan estos valores?
Se est√°n evaluando 10 subconjuntos del dataset (por cv=10).
Cada valor en el array representa el error cuadr√°tico medio negativo (MSE) en una de las validaciones.
Como scoring='neg_mean_squared_error', los valores son negativos, pero en realidad indican qu√© tan lejos est√°n las predicciones de los valores reales.

üîç Interpretaci√≥n:
Valores cercanos a 0 indican un buen rendimiento (menor error).
En este caso, algunos valores son muy cercanos a 0 (-0.0047, -0.0094), lo que sugiere que el modelo predice bastante bien en esas particiones.
Sin embargo, otros valores como -1.8 indican que en algunos conjuntos el error fue mayor.

‚úÖ Conclusi√≥n:
El modelo tiene un buen desempe√±o en la mayor√≠a de los casos, pero hay algunas variaciones dependiendo del conjunto de prueba. Podr√≠a ser √∫til ajustar los hiperpar√°metros con GridSearchCV para mejorar la estabilidad del modelo. üöÄ

## üìù C√≥digo:

````python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(modelo_rf, X, Y, cv=10, scoring='neg_mean_squared_error')
scores
````

## üìâ C√°lculo del RMSE (Ra√≠z del Error Cuadr√°tico Medio)

Para interpretar mejor los errores obtenidos en la validaci√≥n cruzada, calculamos el **RMSE** (*Root Mean Squared Error*), que nos da una m√©trica m√°s intuitiva de la magnitud del error en las predicciones.

üîç Interpretaci√≥n:
El RMSE mide cu√°nto se desv√≠a la predicci√≥n del modelo en promedio.
Valores m√°s bajos indican mejor precisi√≥n. Aqu√≠, algunos valores son cercanos a 0, lo que indica buenas predicciones.
Sin embargo, hay algunos valores como 1.34, lo que sugiere que en algunos casos el error fue m√°s alto.

‚úÖ Conclusi√≥n:
El modelo predice con baja desviaci√≥n en la mayor√≠a de los casos, pero podr√≠a mejorarse con ajustes en los hiperpar√°metros (como n_estimators, max_depth, etc.). üöÄ


````python
import numpy as np
rmse_scores=np.sqrt(-scores)
rmse_scores
````

## üìä An√°lisis de Resultados del RMSE

Para evaluar la estabilidad del modelo, analizamos los puntajes obtenidos en la validaci√≥n cruzada. Calculamos el **promedio y la desviaci√≥n est√°ndar** del **RMSE** para entender la variabilidad del error en las predicciones.

üîç Interpretaci√≥n:
El promedio de RMSE es 0.361, lo que indica que, en general, el modelo tiene una baja desviaci√≥n en sus predicciones. ‚úÖ
La desviaci√≥n est√°ndar es 0.463, lo que significa que hay cierta variabilidad en los errores entre los distintos conjuntos de validaci√≥n. üìâ

‚úÖ Conclusi√≥n:
El modelo es bastante preciso en la mayor√≠a de los casos, pero podr√≠a beneficiarse de un ajuste fino (fine-tuning) para reducir la variabilidad en las predicciones. üöÄ

## üìù C√≥digo:

````python
def mostrar(scores):
  print("Puntajes de RMSE en fold", scores)
  print("promedio de RMSE: ",np.mean(scores))
  print("Desviacion estandar de RMSE: ",np.std(scores))
mostrar(rmse_scores)
````

# üìä Matriz de Confusi√≥n: Explicaci√≥n y Uso  

## üìå ¬øQu√© es una Matriz de Confusi√≥n?  
La **matriz de confusi√≥n** es una herramienta fundamental para evaluar el rendimiento de modelos de clasificaci√≥n. Muestra el n√∫mero de aciertos y errores del modelo, organizados en una tabla que compara las predicciones con los valores reales.  

## üìã Estructura de la Matriz  
Se organiza en una tabla de 2x2 para un problema de clasificaci√≥n binaria (positivo/negativo) o m√°s grande para m√∫ltiples clases.  

|   | Predicci√≥n Positiva | Predicci√≥n Negativa |
|---|--------------------|--------------------|
| **Real Positiva**  | Verdaderos Positivos (**TP**) | Falsos Negativos (**FN**) |
| **Real Negativa**  | Falsos Positivos (**FP**) | Verdaderos Negativos (**TN**) |

### üîπ **Explicaci√≥n de los t√©rminos:**  
- **TP (True Positive)** ‚Üí Predicciones correctas de la clase positiva.  
- **TN (True Negative)** ‚Üí Predicciones correctas de la clase negativa.  
- **FP (False Positive)** ‚Üí Predicciones incorrectas, el modelo clasific√≥ como positivo algo que era negativo (**Falso Alarma**).  
- **FN (False Negative)** ‚Üí Predicciones incorrectas, el modelo clasific√≥ como negativo algo que era positivo (**Error de Omisi√≥n**).  

## üìä M√©tricas Derivadas  
A partir de la matriz de confusi√≥n, se pueden calcular m√©tricas clave:  

- **Precisi√≥n (Precision):**  
  \[
  \text{Precisi√≥n} = \frac{TP}{TP + FP}
  \]
  Mide cu√°ntas de las predicciones positivas fueron realmente correctas.  

- **Sensibilidad / Recall (True Positive Rate - TPR):**  
  \[
  \text{Recall} = \frac{TP}{TP + FN}
  \]
  Indica qu√© tan bien el modelo detecta la clase positiva.  

- **Especificidad (True Negative Rate - TNR):**  
  \[
  \text{Especificidad} = \frac{TN}{TN + FP}
  \]
  Mide la capacidad del modelo para identificar correctamente la clase negativa.  

- **Exactitud (Accuracy):**  
  \[
  \text{Exactitud} = \frac{TP + TN}{TP + TN + FP + FN}
  \]
  Representa el porcentaje total de predicciones correctas del modelo.  

## üìå Importancia en Machine Learning  
La matriz de confusi√≥n es esencial en **modelos de clasificaci√≥n** para entender d√≥nde est√° fallando el modelo y si se requiere balancear clases, ajustar hiperpar√°metros o mejorar los datos de entrenamiento.  

üîπ **Ejemplo:** Si un modelo de detecci√≥n de fraudes predice muchas transacciones leg√≠timas como fraudulentas (**FP alto**), podr√≠a causar problemas en la experiencia del usuario.  

---

‚úÖ **Conclusi√≥n:**  
La matriz de confusi√≥n permite evaluar el desempe√±o del modelo m√°s all√° de la simple exactitud, proporcionando informaci√≥n detallada sobre los errores y aciertos en cada categor√≠a.
