# 🔄 Codificación de Variables Categóricas  

Este código utiliza `LabelEncoder` para convertir variables categóricas en valores numéricos, facilitando su uso en modelos de Machine Learning.  

## 📝 Código:

````python
from sklearn.preprocessing import LabelEncoder

# Crear una instancia de LabelEncoder
le = LabelEncoder()

# Codificar las variables categóricas de tipo 'object' (como 'Gender', 'family_history_with_overweight', etc.)
categorical_columns = ['Gender', 'family_history_with_overweight', 'FAVC', 'CAEC', 'SMOKE', 'SCC', 'CALC', 'MTRANS']

for col in categorical_columns:
    datos[col] = le.fit_transform(datos[col])

# Codificar la columna 'NObeyesdad' (variable objetivo) a números
datos['NObeyesdad'] = le.fit_transform(datos['NObeyesdad'])
````

# ✂️ Separación de Variables: Características (X) y Objetivo (Y)

En este paso, dividimos el conjunto de datos en **variables predictoras (X)** y **variable objetivo (Y)**.

## 📝 Código:

````python
X = datos.drop('NObeyesdad', axis=1)
Y = datos['NObeyesdad']
````

# 🔀 División del Conjunto de Datos: Entrenamiento y Prueba

En este paso, se divide el dataset en dos partes:  
- **80%** para entrenamiento (X_entrena, Y_entrena) 📚

## 📝 Código:

````python
from sklearn.model_selection import train_test_split
X_entrena, X_prueba, Y_entrena, Y_prueba = train_test_split(X, Y, test_size=0.2, random_state=42)
````

# 🌲 Entrenamiento del Modelo con Random Forest

Se utiliza **Random Forest**, un modelo de clasificación basado en múltiples árboles de decisión, para predecir el nivel de obesidad.

📌 Analogía:
Puedes imaginar Random Forest como una votación en la que 100 personas (árboles de decisión) dan su opinión sobre el resultado.

🔹 Ejemplo: Si estás decidiendo qué película ver y le preguntas a 100 amigos, algunos dirán "acción", otros "comedia", pero la decisión final será la que tenga más votos.

✅ En el modelo, cada árbol hace una predicción y la decisión final se toma por mayoría de votos (clasificación) o por promedio (regresión).

## 📝 Código:

````python
from sklearn.ensemble import RandomForestClassifier

modelo_rf = RandomForestClassifier(n_estimators=100, random_state=42)

modelo_rf.fit(X_entrena, Y_entrena)
````

## 📊 Validación Cruzada con `cross_val_score`

Para evaluar la precisión del modelo, utilizamos **validación cruzada** con `cross_val_score`.

🤔 ¿Qué significan estos valores?
Se están evaluando 10 subconjuntos del dataset (por cv=10).
Cada valor en el array representa el error cuadrático medio negativo (MSE) en una de las validaciones.
Como scoring='neg_mean_squared_error', los valores son negativos, pero en realidad indican qué tan lejos están las predicciones de los valores reales.

🔍 Interpretación:
Valores cercanos a 0 indican un buen rendimiento (menor error).
En este caso, algunos valores son muy cercanos a 0 (-0.0047, -0.0094), lo que sugiere que el modelo predice bastante bien en esas particiones.
Sin embargo, otros valores como -1.8 indican que en algunos conjuntos el error fue mayor.

✅ Conclusión:
El modelo tiene un buen desempeño en la mayoría de los casos, pero hay algunas variaciones dependiendo del conjunto de prueba. Podría ser útil ajustar los hiperparámetros con GridSearchCV para mejorar la estabilidad del modelo. 🚀

## 📝 Código:

````python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(modelo_rf, X, Y, cv=10, scoring='neg_mean_squared_error')
scores
````

## 📉 Cálculo del RMSE (Raíz del Error Cuadrático Medio)

Para interpretar mejor los errores obtenidos en la validación cruzada, calculamos el **RMSE** (*Root Mean Squared Error*), que nos da una métrica más intuitiva de la magnitud del error en las predicciones.

🔍 Interpretación:
El RMSE mide cuánto se desvía la predicción del modelo en promedio.
Valores más bajos indican mejor precisión. Aquí, algunos valores son cercanos a 0, lo que indica buenas predicciones.
Sin embargo, hay algunos valores como 1.34, lo que sugiere que en algunos casos el error fue más alto.

✅ Conclusión:
El modelo predice con baja desviación en la mayoría de los casos, pero podría mejorarse con ajustes en los hiperparámetros (como n_estimators, max_depth, etc.). 🚀


````python
import numpy as np
rmse_scores=np.sqrt(-scores)
rmse_scores
````

## 📊 Análisis de Resultados del RMSE

Para evaluar la estabilidad del modelo, analizamos los puntajes obtenidos en la validación cruzada. Calculamos el **promedio y la desviación estándar** del **RMSE** para entender la variabilidad del error en las predicciones.

🔍 Interpretación:
El promedio de RMSE es 0.361, lo que indica que, en general, el modelo tiene una baja desviación en sus predicciones. ✅
La desviación estándar es 0.463, lo que significa que hay cierta variabilidad en los errores entre los distintos conjuntos de validación. 📉

✅ Conclusión:
El modelo es bastante preciso en la mayoría de los casos, pero podría beneficiarse de un ajuste fino (fine-tuning) para reducir la variabilidad en las predicciones. 🚀

## 📝 Código:

````python
def mostrar(scores):
  print("Puntajes de RMSE en fold", scores)
  print("promedio de RMSE: ",np.mean(scores))
  print("Desviacion estandar de RMSE: ",np.std(scores))
mostrar(rmse_scores)
````

## ⚖️ Normalización de Datos con StandardScaler  

Para mejorar el rendimiento del modelo, aplicamos **normalización** a los datos utilizando `StandardScaler`. 📏  

🔍 ¿Qué hace este paso?

StandardScaler transforma los datos para que tengan una media de 0 y una desviación estándar de 1. 📊

Estandariza las variables, asegurando que cada una tenga el mismo peso en el modelo. ⚖️

## 🔍 Búsqueda de Hiperparámetros con GridSearchCV  

Para mejorar el rendimiento del modelo **Random Forest**, usamos `GridSearchCV` para encontrar los mejores hiperparámetros. 🚀

🔍 ¿Qué hace este paso?

Explora diferentes combinaciones de n_estimators, min_samples_split y max_depth. 📊

Usa validación cruzada (cv=10) para evaluar cada combinación. 🔄

Optimiza el modelo buscando minimizar el error cuadrático medio negativo (neg_mean_squared_error). ⚖️

Encuentra la mejor configuración de hiperparámetros. 🎯

### 📝 Código  

````python
from sklearn.model_selection import GridSearchCV
#forest_reg = RandomForestRegressor()
# Definimos los posibles parámetros
n_estimators = [3, 10, 30, 50, 80, 100]
min_samples_split = [2, 4, 6, 8]
max_depth = [2, 4, 6, 8, 10]

params = {
    'n_estimators': n_estimators,
    'min_samples_split': min_samples_split,
    'max_depth': max_depth
}


#Aplicamos el método de ajuste

grid_search = GridSearchCV(estimator=modelo_rf, param_grid=params, cv=10, scoring='neg_mean_squared_error')
grid_search.fit(X_prueba_scaled, Y_prueba)
````

## 🏆 Selección de los Mejores Hiperparámetros  

Después de aplicar `GridSearchCV`, obtenemos los **mejores hiperparámetros** para nuestro modelo **Random Forest**.  

````python
print("Mejores parámetros:", grid_search.best_params_)
````

## 🚀 Creación del Modelo Optimizado con Random Forest Regressor  

Después de encontrar los mejores hiperparámetros con `GridSearchCV`, creamos un **nuevo modelo** de **Random Forest Regressor** con estos valores óptimos.

# 📊 Matriz de Confusión: Explicación y Uso  

## 📌 ¿Qué es una Matriz de Confusión?  
La **matriz de confusión** es una herramienta fundamental para evaluar el rendimiento de modelos de clasificación. Muestra el número de aciertos y errores del modelo, organizados en una tabla que compara las predicciones con los valores reales.  

## 📋 Estructura de la Matriz  
Se organiza en una tabla de 2x2 para un problema de clasificación binaria (positivo/negativo) o más grande para múltiples clases.  

|   | Predicción Positiva | Predicción Negativa |
|---|--------------------|--------------------|
| **Real Positiva**  | Verdaderos Positivos (**TP**) | Falsos Negativos (**FN**) |
| **Real Negativa**  | Falsos Positivos (**FP**) | Verdaderos Negativos (**TN**) |

### 🔹 **Explicación de los términos:**  
- **TP (True Positive)** → Predicciones correctas de la clase positiva.  
- **TN (True Negative)** → Predicciones correctas de la clase negativa.  
- **FP (False Positive)** → Predicciones incorrectas, el modelo clasificó como positivo algo que era negativo (**Falso Alarma**).  
- **FN (False Negative)** → Predicciones incorrectas, el modelo clasificó como negativo algo que era positivo (**Error de Omisión**).  

## 📊 Métricas Derivadas  
A partir de la matriz de confusión, se pueden calcular métricas clave:  

- **Precisión (Precision):**  
  \[
  \text{Precisión} = \frac{TP}{TP + FP}
  \]
  Mide cuántas de las predicciones positivas fueron realmente correctas.  

- **Sensibilidad / Recall (True Positive Rate - TPR):**  
  \[
  \text{Recall} = \frac{TP}{TP + FN}
  \]
  Indica qué tan bien el modelo detecta la clase positiva.  

- **Especificidad (True Negative Rate - TNR):**  
  \[
  \text{Especificidad} = \frac{TN}{TN + FP}
  \]
  Mide la capacidad del modelo para identificar correctamente la clase negativa.  

- **Exactitud (Accuracy):**  
  \[
  \text{Exactitud} = \frac{TP + TN}{TP + TN + FP + FN}
  \]
  Representa el porcentaje total de predicciones correctas del modelo.  

## 📌 Importancia en Machine Learning  
La matriz de confusión es esencial en **modelos de clasificación** para entender dónde está fallando el modelo y si se requiere balancear clases, ajustar hiperparámetros o mejorar los datos de entrenamiento.  

🔹 **Ejemplo:** Si un modelo de detección de fraudes predice muchas transacciones legítimas como fraudulentas (**FP alto**), podría causar problemas en la experiencia del usuario.  

---

✅ **Conclusión:**  
La matriz de confusión permite evaluar el desempeño del modelo más allá de la simple exactitud, proporcionando información detallada sobre los errores y aciertos en cada categoría.
